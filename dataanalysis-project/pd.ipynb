```python
import pandas as pd

s = pd.Series([1, 2, 3, 4, 5],index=['a', 'b', 'c', 'd', 'e'])
```

```python
print(s.index)
print(s.values)
print(s.shape,s.ndim,s.size,s.dtype)
print(s.loc['a':'c'])
print(s.iloc[0:2])
print(s.at['a'])#不能切片
print(s.iat[0])#同上
```

```python
#print(s[1])#最好不要这么写,最好用标签
print(s['a'])
print(s[s>2])
print(s.head())#前5行 tail()后5行 加参数可以控制几行
```

```python
import numpy as np
ss = pd.Series([10,2,np.nan,None,3,4,5],index=['A','B','C','D','E','F','G'])
print(ss)
```

```python
ss.head()
ss.tail(2)
```

```python
ss.describe()
```

```python
print(ss.count())
print(ss.sum())
print(ss.mean())
print(ss.std())
print(ss.var())
print(ss.min())
print(ss.max())
print(ss.median())
print(ss.quantile(0.33))#分位数
print(ss.mode())#众数
print(ss.keys())
print(ss.index)
```

```python
print(ss.isna())
print(ss.isin([1,2,3]))
```

```python
ss.sort_values()#参数ascending=False控制升降序
ss.sort_index()
print(ss.value_counts())
ss.drop_duplicates()#去重
ss.unique()#也是去重，只是上面返回值还是series
```

```python
np.random.seed(42)
scores = pd.Series(np.random.randint(50,101,10),index=['学生'+str(i) for i in range(1,11)])
print(scores)
```

```python
print(scores.mean())
print(scores.max())
print(scores.min())
print(scores.diff())#计算变化值
scores[scores>scores.mean()]
```

```python
sales = pd.Series([120,135,145,160,155,170,180,175,190,200,210,220],index=pd.date_range("2022-01-01",periods=12,freq="MS"))
sales
```

```python
sales.resample("QS").mean()
```

```python
sales.pct_change()
```

```python
b = sales.pct_change()>0
b[b.rolling(3).sum()==3].keys().tolist()
#.between_time 筛选一定时间内的series(一天之内的)
```
